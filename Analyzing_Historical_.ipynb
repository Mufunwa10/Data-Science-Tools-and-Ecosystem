## Question 1

```python
import yfinance as yf
import pandas as pd

# Download historical data for a stock
tsla = yf.Ticker("TSLA")
tesla_data = tsla.history(period="max")

# Reset the index
tesla_data.reset_index

# Save the dataframe 
tesla_data.to_csv('tesla_data.csv', index=False)

# drop empty columns

tesla_data.drop(columns=["Dividends","Stock Splits"],inplace=True)

# Display the first five rows
tesla_data.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

import warnings
# Ignore all warnings
warnings.filterwarnings("ignore", category=FutureWarning)

# URL of the webpage
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"

# Send an HTTP request to the webpage
response = requests.get(url)
html_data = response.text

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_data, 'html.parser')

# Find the table containing Tesla revenue data
table = soup.find_all('table')[0] 

# Convert the HTML table into a DataFrame
gme_revenue = pd.read_html(str(table))[0]

# Display the last five rows of the dataframe
gme_revenue.tail()
```

## Question 5

```python
import yfinance as yf
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(10, 5))
    plt.plot(data.index, data['Close'])
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price USD')
    plt.grid(True)
    plt.show()

# Use the Ticker module to get Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Graph the Tesla stock data
make_graph(tesla_data, 'Tesla Stock Data')

```
